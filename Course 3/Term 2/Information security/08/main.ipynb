```python
def linear_congruential_generator(a, c, n, seed):
    while True:
        seed = (a * seed + c) % n
        yield seed

a = 430
c = 2531
n = 11979
seed = 1

generator = linear_congruential_generator(a, c, n, seed)

for _ in range(10):
    print(next(generator))
```

```python
import time

def RC4(key, text):
    S = list(range(256))
    j = 0
    out = []

    for i in range(256):
        j = (j + S[i] + key[i % len(key)]) % 256
        S[i], S[j] = S[j], S[i]

    i = j = 0
    for char in text:
        i = (i + 1) % 256
        j = (j + S[i]) % 256
        S[i], S[j] = S[j], S[i]
        out.append(chr(char ^ S[(S[i] + S[j]) % 256]))

    return ''.join(out)

# Ключ в виде десятичных чисел
key = [1, 11, 21, 31, 41, 51]
key = [k % 256 for k in key]  # Преобразование ключа в байты

# Текст для шифрования
text = "This is a test message."

# Преобразование текста в байты
text = [ord(c) for c in text]

start_time = time.time()
cipher_text = RC4(key, text)
end_time = time.time()

print("Cipher text:", cipher_text)
print("Time taken:", end_time - start_time)
```
