```python
import time
from sympy import mod_inverse

def encrypt_decrypt():
    def Sverhvozr(start, step, length):
        arr = [step]
        for i in range(start, length):
            sum = 0
            for el in arr:
                sum += el
            arr.append(sum + step)
        return arr

    secretKey = Sverhvozr(5, 5, 8)

    def Normvozr(secretKey):
        sum = 0
        for el in secretKey:
            sum += el
        n = sum + 5
        a = n - 1
        arr = [(element * a) % n for element in secretKey]
        return arr[::-1]

    message = "hamsterGlavnyi123"
    # Секретный ключ (сверхвозрастающая последовательность)
    w = [2, 7, 11, 21, 42, 89, 180, 354]
    # Модуль (больше суммы элементов ключа)
    q = 881
    # Множитель (меньше модуля и взаимно простой с модулем)
    r = 588
    b = []
    S = 0
    cipher = []
    S1 = []
    text = ""

    start = time.time()
    # Создание открытого ключа
    b = [(element * r) % q for element in w]

    # Преобразование каждого символа сообщения в двоичный код и шифрование
    for i in range(len(message)):
        char = list(format(ord(message[i]), '08b'))
        for i in range(len(char)):
            S += int(char[i]) * b[i]
        cipher.append(S)
        S = 0

    end = time.time()
    executionTime = end - start

    encrypted_text = ' '.join(map(str, cipher))

    start = time.time()

    # Вычисление обратного элемента к множителю по модулю
    n = mod_inverse(r, q)

    # Преобразование шифртекста обратно в двоичный код
    S1 = [(element * n) % q for element in cipher]

    # Преобразование двоичного кода обратно в текст
    S2 = sorted(w, reverse=True)
    for element in S1:
        n = element
        m = [0, 0, 0, 0, 0, 0, 0, 0]
        while n != 0:
            i = 0
            f = S2[i]
            while n < f:
                i += 1
                f = S2[i]
            m[w.index(f)] = 1
            n = n - f
        binCode = [chr(int(''.join(map(str, m)), 2))]
        text += ''.join(binCode)

    end = time.time()
    executionTime = end - start


    result = f"Время зашифрования: {executionTime}с\n"
    result += f"Зашифрованный текст: {encrypted_text}\n"
    result += f"Время расшифрования: {executionTime}с\n"
    result += f"Расшифрованный текст: {text}\n"
    result += f"Сверхвозрастающая последовательность: {Sverhvozr(5, 5, 8)}\n"
    result += f"Нормальная последовательность: {Normvozr(secretKey)}\n"

    print(result)


encrypt_decrypt()
```
